```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_total = pd.read_excel("data.xlsx",sheet_name='total')
df_total
```

```python


fig, ax = plt.subplots()
plt.grid(False)
color1 = '#4682B4'
ax.bar(df_total['year'], df_total['increase_of_population'], color=color1,label='Population Increase')


ax2 = plt.twinx()
ax2.plot(df_total['year'], df_total
['total_population'],color='orange', label='Total Population', linewidth=5)


plt.title('Population Increase')
ax.set_xlabel('Year')

# y labels and ticks
ax.set_ylabel('Population Increase / Year',color=color1)
ax.tick_params(axis='y', colors=color1)
ax2.set_ylabel('Total Population',color='orange')
ax2.tick_params(axis='y', colors='orange')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
```

```python
# stack
fig,ax = plt.subplots()
x=df_female_bearing_age['year']
ax.bar(x,df_female_bearing_age['15-19__'])
ax.bar(x,df_female_bearing_age['20-24__'],bottom=df_female_bearing_age['15-19__'])
ax.bar(x,df_female_bearing_age['25-29__'],bottom=df_female_bearing_age['15-19__']+df_female_bearing_age['20-24__'])
ax.bar(x,df_female_bearing_age['30-34__'],bottom=df_female_bearing_age['15-19__']+df_female_bearing_age['20-24__']+df_female_bearing_age['25-29__'])
ax.bar(x,df_female_bearing_age['35-39__'],bottom=df_female_bearing_age['15-19__']+df_female_bearing_age['20-24__']+df_female_bearing_age['25-29__']+df_female_bearing_age['30-34__'])
ax.bar(x,df_female_bearing_age['40-44__'],bottom=df_female_bearing_age['15-19__']+df_female_bearing_age['20-24__']+df_female_bearing_age['25-29__']+df_female_bearing_age['30-34__']+df_female_bearing_age['35-39__'])
ax.bar(x,df_female_bearing_age['45-49__'],bottom=df_female_bearing_age['15-19__']+df_female_bearing_age['20-24__']+df_female_bearing_age['25-29__']+df_female_bearing_age['30-34__']+df_female_bearing_age['35-39__']+df_female_bearing_age['40-44__'])

# 添加标题和标签
plt.title('Stacked Area Chart of Population by Age Group')
plt.xlabel('Year')
plt.ylabel('Population')

# 添加图例
plt.legend(loc='upper left')  # 可根据需要调整图例位置
```

```python
df_correlation = pd.read_excel('data.xlsx',sheet_name='correlation')
df_correlation
```

```python
columns_of_interest = ['total_population', 'increase_of_population', 'child_bearing_age_population','bearing_percen_in_female','death_rate','total_births','total_fertility_rate','net_migration','marriages','divorces'] 
correlation_matrix = df_correlation[columns_of_interest].corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
```
